```python
import os
import pandas as pd
import torch
import matplotlib.pyplot as plt
from collections import defaultdict

# =========================================================
# CONFIG
# =========================================================

DATASET_ROOT = "data"   # because your cwd = ecg_afib_detection

# =========================================================
# HELPERS
# =========================================================

def load_pt(path):
    return torch.load(path, map_location="cpu")

def describe_tensor(t):
    X = t["X"]
    return {
        "samples": X.shape[0],
        "leads": X.shape[1],
        "timesteps": X.shape[2],
        "shape": tuple(X.shape),
        "afib_ratio": float(t["y"].float().mean()),
        "patients": len(set(t["patient_ids"]))
    }

# =========================================================
# SANITY + DUPLICATE CHECKS
# =========================================================

def run_sanity_checks(df, mode_col):
    print("\nSanity checks:")
    ok = True

    # -------- Patient leakage --------
    g = df.groupby("patient_id")[mode_col].nunique()
    if (g > 1).any():
        print("✖ Patient leakage detected")
        ok = False
    else:
        print("✔ No patient leakage")

    # -------- Record leakage --------
    g = df.groupby("record_id")[mode_col].nunique()
    if (g > 1).any():
        print("✖ Record leakage detected")
        ok = False
    else:
        print("✔ No record leakage")

    # -------- Segment leakage --------
    g = df.groupby(["record_id", "segment_index"])[mode_col].nunique()
    if (g > 1).any():
        print("✖ Segment leakage detected")
        ok = False
    else:
        print("✔ No segment leakage")

    # -------- Label consistency --------
    g = df.groupby("record_id")["label"].nunique()
    if (g > 1).any():
        print("✖ Label conflicts per record")
        ok = False
    else:
        print("✔ Labels consistent per record")

    # -------- Duplicate segments --------
    dup_cols = ["patient_id", "record_id", "segment_index"]
    n_dups = df.duplicated(subset=dup_cols).sum()

    if n_dups > 0:
        print(f"✖ Duplicate segments found: {n_dups}")
        ok = False
    else:
        print("✔ No duplicate segments")

    return ok, n_dups

# =========================================================
# MAIN DATASET CHECK
# =========================================================

def check_dataset(dataset_path, dataset_name):

    print("\n" + "=" * 100)
    print(f"DATASET: {dataset_name}")
    print("=" * 100)

    for fs_dir in sorted(os.listdir(dataset_path)):
        if not fs_dir.endswith("hz"):
            continue

        fs_path = os.path.join(dataset_path, fs_dir)
        if not os.path.isdir(fs_path):
            continue

        print(f"\nSampling rate folder: {fs_dir}")
        print("-" * 80)

        # ---------------- CSV ----------------
        csv_file = next(
            (f for f in os.listdir(fs_path) if f.startswith("samples_")), None
        )
        if csv_file is None:
            print("⚠ No CSV found")
            continue

        df = pd.read_csv(os.path.join(fs_path, csv_file))
        mode_col = "fold" if "fold" in df.columns else "split"
        mode_name = "FOLD" if mode_col == "fold" else "SPLIT"

        print(f"Mode            : {mode_name}")
        print(f"Total samples   : {len(df)}")
        print(f"Label counts    : {df['label'].value_counts().to_dict()}")
        print(f"Sampling rate   : {df['fs'].unique().tolist()}")
        print(f"Unique patients : {df['patient_id'].nunique()}")
        print(f"Unique records  : {df['record_id'].nunique()}")

        # ---------------- TENSORS ----------------
        tensors = {}

        if mode_col == "fold":
            data_pt = os.path.join(fs_path, "data.pt")
            if os.path.exists(data_pt):
                tensors["CV (folds)"] = load_pt(data_pt)
        else:
            for split in ["train", "val", "test"]:
                p = os.path.join(fs_path, f"{split}.pt")
                if os.path.exists(p):
                    tensors[split] = load_pt(p)

        # Hold-out test (optional)
        holdout_pt = os.path.join(fs_path, "test", "test.pt")
        if os.path.exists(holdout_pt):
            tensors["HOLD-OUT test"] = load_pt(holdout_pt)

        # ---------------- Tensor info ----------------
        print("\nTensor info:")
        for name, t in tensors.items():
            info = describe_tensor(t)
            print(f"{name}")
            print(f"  Shape        : {info['shape']}")
            print(f"  Samples      : {info['samples']}")
            print(f"  Patients     : {info['patients']}")
            print(f"  AFIB ratio   : {info['afib_ratio']:.3f}")

        # ---------------- Sanity checks ----------------
        ok, n_dups = run_sanity_checks(df, mode_col)

        # ---------------- PLOTS ----------------
        fig, axes = plt.subplots(1, 3, figsize=(16, 4))

        # Label distribution
        df["label"].value_counts().sort_index().plot(
            kind="bar", ax=axes[0], title="Label Distribution"
        )
        axes[0].set_xlabel("label")
        axes[0].set_ylabel("samples")

        # Fold / split distribution
        df[mode_col].value_counts().sort_index().plot(
            kind="bar", ax=axes[1], title=f"{mode_col} Distribution"
        )
        axes[1].set_xlabel(mode_col)
        axes[1].set_ylabel("samples")

        # Duplicate visualization
        if n_dups > 0:
            df[df.duplicated(
                subset=["patient_id", "record_id", "segment_index"],
                keep=False
            )][mode_col].value_counts().plot(
                kind="bar", ax=axes[2], title="Duplicates per fold/split"
            )
        else:
            axes[2].text(0.5, 0.5, "No duplicates",
                         ha="center", va="center", fontsize=12)
            axes[2].set_axis_off()

        plt.tight_layout()
        plt.show()

# =========================================================
# RUN ALL DATASETS
# =========================================================

for dataset_name in sorted(os.listdir(DATASET_ROOT)):
    path = os.path.join(DATASET_ROOT, dataset_name)
    if os.path.isdir(path):
        check_dataset(path, dataset_name)
```
